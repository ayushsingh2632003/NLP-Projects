```python
import pandas as pd
from gensim.models import Word2Vec
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, classification_report
from nltk.tokenize import word_tokenize

# Assuming your data is stored in a DataFrame named 'df'
# with columns 'reviews' and 'sentiments'
# Example: df = pd.read_csv('your_dataset.csv')

# Convert the 'reviews' column to lowercase and tokenize each review
df['reviews'] = df['reviews'].apply(lambda review: ' '.join(review).lower())
df['reviews'] = df['reviews'].apply(lambda review: word_tokenize(review))

word2vec_model = Word2Vec(sentences=df['reviews'], vector_size=100, window=5, min_count=1, workers=4)

# Feature Extraction
def get_feature_vector(review):
    word_vectors = [word2vec_model.wv[word] for word in review if word in word2vec_model.wv]
    if not word_vectors:
        return None
    return sum(word_vectors) / len(word_vectors)

df['feature_vector'] = df['reviews'].apply(get_feature_vector)

# Remove rows with None feature vectors
df = df.dropna()

# Data Splitting
X_train, X_test, y_train, y_test = train_test_split(df['feature_vector'].tolist(), df['sentiments'], test_size=0.2, random_state=42)

# Hyperparameter Tuning using Grid Search
param_grid = {'C': [0.1, 1, 10, 100], 'kernel': ['linear', 'rbf']}
grid_search = GridSearchCV(SVC(), param_grid, cv=5)
grid_search.fit(X_train, y_train)

best_svm_model = grid_search.best_estimator_

# Evaluate the best model
y_pred_best = best_svm_model.predict(X_test)
accuracy_best = accuracy_score(y_test, y_pred_best)
report_best = classification_report(y_test, y_pred_best)

print(f'Best Model Accuracy: {accuracy_best}')
print(f'Best Model Classification Report:\n{report_best}')
```

```python
# Define a function for text preprocessing and feature extraction
def preprocess_and_extract_features(text):
    # Preprocess the input text
    text = ' '.join(word_tokenize(text.lower()))

    # Extract features using the trained Word2Vec model
    feature_vector = get_feature_vector(text)

    return feature_vector

# Example of a random text
random_text = "This product is amazing! I love it."

# Preprocess and extract features for the random text
random_text_features = preprocess_and_extract_features(random_text)

# Check sentiment using the trained SVM model
predicted_sentiment = best_svm_model.predict([random_text_features])[0]

# Map the predicted sentiment to "good" or "bad"
sentiment_label = "good" if predicted_sentiment == 1 else "bad"

# Print the result
print(f"Predicted Sentiment for the random text: {sentiment_label}")
```
